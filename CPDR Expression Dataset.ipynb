```python
data_input=data_input.drop(['Genbank','Probeset ID'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
data_scaled=scaler.fit_transform(data_input)
```

```python
ndata_input=pd.read_csv('C:/Users/Sai/Desktop/Genetic A/normal pd dataset.txt',sep='\t')
```

```python
ndata_input
ndata_input=ndata_input.drop(['Genbank','ID'],axis=1)
```

```python
ndata_input=ndata_input.dropna()
```

```python
ndata_scaled=scaler.fit_transform(ndata_input)
```

```python
ndata_scaled_df=pd.DataFrame(ndata_scaled,index=ndata_input.index,columns=ndata_input.columns)
```

```python
ndata_scaled_df=ndata_scaled_df.T
```

```python
ndata_scaled_df
```

```python
ndata_scaled_target=ndata_scaled_df.join(target)
ndata_scaled_target['Target']=np.zeros((20,1))
```

```python
ndata_scaled_target
```

```python
import random
X=pd.concat([data_scaled_df,ndata_scaled_df])
```

```python
y=pd.concat([data_scaled_target['Target'],ndata_scaled_target['Target']])
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=5)
X_pca=pca.fit_transform(X)

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X_pca,y,test_size=0.5,random_state=172)
```

```python
from sklearn.svm import SVC
```

```python
svm_clf=SVC()
```

```python
svm_clf.fit(X_train,y_train)
```

```python
svm_clf.score(X_train,y_train)
```

```python
from sklearn.linear_model import LogisticRegression
lg_clf=LogisticRegression(random_state=69)
lg_clf.fit(X_train,y_train)
```

```python
lg_clf.score(X_test,y_test)
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=2,random_state=69)
```

```python
kmeans.fit(X_train,y_train)
y_kmeans=kmeans.predict(X_train)
print(y_kmeans)
print(y_train==y_kmeans)
plt.scatter(X_train.values[:,0],X_train.values[:,1],c=y_kmeans[0],s=50,cmap='viridis')
centers=kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='black',s=200,alpha=.5)
```
